```python
# Grab y_labels and remove from data
y_labels = data[" Label"]
data = data.drop(labels=" Label", axis=1)
data.columns
```

```python
y_labels.value_counts()
```

```python
data_cleaned = data.apply(lambda x: x.fillna(x.mean()),axis=0)
nan_rows = data_cleaned[data_cleaned.isnull().any(1)]
nan_rows
```

```python
#one hot encode labels
y_labels_as_int_list = []
for label in y_labels:
    if label == "Infiltration":
        y_labels_as_int_list.append(1)
    else:
        y_labels_as_int_list.append(0)
y_labels_encoded = to_categorical(y_labels_as_int_list)

```

```python
nn_input_dim = data.shape[1]
nn_input_dim
```

```python
neural_network = Sequential()
neural_network.add(Dense(128, input_dim=nn_input_dim, activation="relu"))
neural_network.add(BatchNormalization())
neural_network.add(Dense(256, activation="relu"))
neural_network.add(BatchNormalization())
neural_network.add(Dense(512, activation="relu"))
neural_network.add(BatchNormalization())
neural_network.add(Dense(2))
neural_network.add(Activation(tf.nn.softmax))
```

```python
neural_network.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
neural_network.summary()
```

```python
neural_network.fit(data, y_labels_encoded, epochs=100, batch_size=100)
```

```python
incorrects = np.nonzero(neural_network.predict(data).reshape((-1,)) != y_labels_encoded)
```
